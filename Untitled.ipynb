```python
net = Pix2Code().cuda()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr = 0.0001)
```

```python
for epoch in range(10):
    net.zero_grad()
    for j, i in enumerate(dataset):
        a, b, c = i
        a = a.float().cuda()
        b = b.float().cuda()
        c = c.float().cuda()
        output = net(a, b)
#         print(output.shape)
#         print(c.shape)
#         output = torch.argmax(output, 2)
        output = output.squeeze(0)
        c = c.squeeze(0)
        c = torch.argmax(c, 1)
#         print(c.shape)
#         print(output.shape)
        loss = criterion(output, c)
        loss.backward()
        if j%10 == 0:
            optimizer.step()
            print('Loss: {}, Epoch: {}'.format(loss.data, epoch))
            net.zero_grad()
#         break
```
